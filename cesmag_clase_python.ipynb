```python
alfabeto = 'abcdefghijklmnñopqrstuvwxyz12345'
print(len(alfabeto))
mensaje_original = input("Dame un Mensaje a cifrar: ")

clave = 16

# cifrar mensaje con rot13
mensaje_cifrado = ''
for caracter in mensaje_original:
    if caracter in alfabeto:
        posicion = alfabeto.find(caracter)
        nueva_posicion = (posicion + clave) % len(alfabeto)
        mensaje_cifrado = mensaje_cifrado + alfabeto[nueva_posicion]
print(f'Mensaje original: {mensaje_original}')
print(f'Mensaje cifrado: {mensaje_cifrado}')
```

```python
import hashlib
data = input("Dame un texto: ")
encoded_data = data.encode('utf-8')
print("encoded_data",encoded_data)
md5_hash = hashlib.md5()
md5_hash.update(encoded_data)
print("md5_hash",md5_hash)
hex_digest = md5_hash.hexdigest()
print(f"Original string: {data}")
print(f"MD5 hash: {hex_digest}")
```

```python
import pandas as pd

ruta_archivo = "/content/compras_semana.csv"

datos = pd.read_csv(ruta_archivo)
datos['total'] = datos['precio'] * datos['cantidad']

total_compras=datos['total'].sum()
print("total de compra: ", total_compras)

promedio_compras=datos['total'].mean()
print("promedio de compra: ", promedio_compras)

mediana_compras = datos['total'].median()
print("mediana de la compra: ", mediana_compras)

compra_minima=datos['total'].min()
print("compra minima: ", compra_minima)

std = datos['total'].std()
print("Desviacion estandar: ", std)

var = datos['total'].var()
print("Varianza: ", var)

registro_conteo = datos['total'].count()
print("Registro conteo: ", registro_conteo)

datos['total'].describe()
```
